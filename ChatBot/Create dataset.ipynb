```python
with open('Jasjeet.txt') as f:
    chatlist=[]
    sublist=[]
    chat=f.readline()
    while(chat!=""):
        
#         print(chat)
        botname="KSHITIJ SRIVASTAVA:"
        myname="Jasjeet Vit:"
        
        if(botname in chat):
            chatlist.append(sublist)
            sublist=[]
            sublist.append(chat[chat.index(botname)+len(botname):])
            print(chat[chat.index(botname)+len(botname):])
        
        if(myname in chat):
            s=chat[chat.index(myname)+len(myname):]
            sublist.append(s)
            print(chat[chat.index(myname)+len(myname):])
        
        chat=f.readline()
        
            
    chatlist=chatlist[1:]
    print(chatlist)
```

```python
with open ('Whats.txt','a') as o:
    for chat in chatlist:
        chatorder=""
        for line in chat:
            chatorder+=line[0:len(line)-1]+","
        
        o.write(chatorder[1:len(chatorder)-1])
        o.write("\n")
```
